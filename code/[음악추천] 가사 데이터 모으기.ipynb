```python
tag_list = ['슬픔','이별','회상','추억','기분전환','휴식','한강','감성','여름','위로','여행','불안','빡칠때','화날때']

data_list = []
cnt = 0
for tag in tag_list:
    temp = pd.read_csv(f'./songDB_{tag}.csv')
    if 'key' in temp.columns:
        temp = temp.drop(columns=['key'])
    temp['tag'] = tag
    cnt += temp.shape[0]
    data_list.append(temp)

songDB = pd.concat(data_list).reset_index(drop=True)                                          # 데이터 통합
songDB['title-singer'] = songDB['title'] + '-' + songDB['singer']                             # id만들기
print(cnt, songDB.shape)
songDB = songDB.dropna(axis=0)                                                                # 결측값 제거
songDB['temp'] = songDB['lyrics'].apply(lambda x: re.sub('[^가-힣]','',x))                    # 영어 가사 탐색
songDB = songDB[songDB['temp']!=''].reset_index(drop=True).drop(columns=['temp'])             # 필터링, temp열 제거

# 데이터 저장
songDB.to_csv('./songDB_origin.csv', index=False)
songDB
```

# **플레이리스트 중복 확인하기**

```python
play_tag = {}
for play, tag in zip(songDB['playlist'],songDB['tag']):
    if play in play_tag.keys():
        if tag not in play_tag[play]:
            play_tag[play].append(tag)
    else:
        play_tag[play] = [tag]

print(play_tag)
```

```python
songDB = songDB.drop_duplicates(['playlist','title-singer']).reset_index(drop=True)     # 중복 제거
songDB['tag'] = songDB['playlist'].apply(lambda x: ', '.join(play_tag[x]))              # 태그를 다시 지정
songDB['final_tag'] = songDB['genre'] + ', ' +songDB['tag']

# 데이터 저장
songDB.to_csv('./songDB.csv', index=False)
songDB
```

# **감정분석할 가사 리스트 만들기**

```python
lyricsDB = songDB[['title-singer','lyrics']]
lyricsDB = lyricsDB.drop_duplicates(['title-singer']).reset_index(drop=True)
lyricsDB
```

```python
# 추론
def lyrics_preprocessing(lyrics):
    lyrics = re.sub('[^가-힣\s]','',lyrics)

    lyrics_str = ''
    for i, x in enumerate(lyrics.split('\n')):
        if x.strip() != '' and x not in lyrics_str:
            if i > 0:
                lyrics_str += ' ' + x.strip()
            else:
                lyrics_str += x.strip()

    result = lyrics_str.strip()
    if result == '':
        return 'X'
    return lyrics_str.strip()
```

```python
# 가사 전처리
lyrics_list = []
for i, temp in enumerate(lyricsDB['lyrics'].tolist()):
    result = lyrics_preprocessing(temp)
    lyrics_list.append(lyrics_preprocessing(temp))

lyricsDB['new_lyrics'] = lyrics_list

# 가사 목록 저장
lyricsDB.to_csv('lyricsDB.csv', index=False)
lyricsDB
```
