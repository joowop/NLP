# **Functions**

```python
from textrank import KeywordSummarizer
from kiwipiepy import Kiwi
import re
from konlpy.tag import Okt

okt = Okt()
kiwi = Kiwi()
kiwi.prepare()
```

```python
# okt
def okt_tokenize(sent):
    sent = re.sub('[^ㄱ-ㅣ가-힣\s]','',sent)
    tokens = okt.pos(sent)
    token_pos = [f'{x[0]}/{x[1]}' for x in tokens if x[1]!='Josa']       # if ('NN' in x.tag or 'VA' in x.tag)

    return token_pos
```

```python
# 명사만 추출
def kiwi_tokenize(sent):
    sent = re.sub('[^ㄱ-ㅣ가-힣\s]','',sent)
    tokens = kiwi.tokenize(sent)
    token_pos = [f'{x.form}/{x.tag}' for x in tokens if 'J' not in x.tag]       # if ('NN' in x.tag or 'VA' in x.tag)

    return token_pos
```

```python
# 스페이스로
def space_tokenize(sent):
    sent = re.sub('[^ㄱ-ㅣ가-힣\s]','',sent)
    tokens = sent.split(' ')

    return tokens
```

```python
# Example
sent = "안녕하세요 저는 오늘 기분이 너무 좋아요."
kiwi_tokenize(sent)
```

```python
def extract_keyword(tokenize, sents):
    sents = sents.split('\n')

    sent_list = []
    for sent in sents:
        if sent != '':
            tokens = kiwi_tokenize(sent.strip())
            sent_list.append(sent)

    keyword_extractor = KeywordSummarizer(
        tokenize = tokenize,
        window = -1,
        verbose = False
    )

    keywords = keyword_extractor.summarize(sents, topk=30)

    return keywords
```
