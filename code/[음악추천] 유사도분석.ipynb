# **1. 데이터 불러오기**

```python
with open('./song2emotion.pkl','rb') as f:
    song2emotion = pickle.load(f)
```

```python
lyrics = pd.read_csv('./lyricsDB_emotion.csv')
print('lyrics',lyrics.shape)

songs = pd.read_csv('./songDB.csv')
songs
```

```python
# emotion 매칭 시키기 + tag 합치기
songs['emotion'] = songs['title-singer'].apply(lambda x: song2emotion[x])
songs['final_tag'] = songs['final_tag'] + ', ' + songs['emotion']
songs.to_csv('songDB_final.csv', index=False)
songs
```

## 1) Version1. 노래 기반

```python
songs_ind = songs.drop_duplicates(['title-singer']).reset_index(drop=True)     # 중복 제거

songs_ind
```

# **2. 플레이리스트 생성**

```python
# 전체 가수-곡에 따른 id 부여 : song_dict
temp = set(songs['title-singer'].tolist())
song2idx = {x:i for i,x in enumerate(temp)}
idx2song = {i:x for i,x in enumerate(temp)}

with open('./song2idx.pkl', 'wb') as f:
    pickle.dump(song2idx,f)
with open('./idx2song.pkl', 'wb') as f:
    pickle.dump(idx2song,f)
```

```python
# songs dataframe에 idx 변수 추가
songs['idx'] = songs['title-singer'].apply(lambda x: song2idx[x])
```

```python
# 각 플레이리스트의 idx를 리스트로 만듦 playlist_idx
playlist = songs['playlist'].unique()
songs_vec = []
for play in playlist:
    temp = songs[songs['playlist']==play]['idx'].tolist()
    songs_vec.append(temp)
```

```python
tag_list = []
for temp in songs['final_tag']:
    temp = temp.split(', ')
    for tag in temp:
        if tag not in tag_list and tag != '-':
            tag_list.append(tag)

tag2idx = {x:i for i,x in enumerate(tag_list)}
idx2tag = {i:x for i,x in enumerate(tag_list)}
```

```python
# 가수-곡에 따른 태그 딕셔너리 생성
song_tag = {}
song_tag_idx = {}
for song, tag in zip(songs_ind['title-singer'],songs_ind['final_tag']):
    temp = tag.split(', ')
    song_tag[song] = temp
    temp_idx = [tag2idx[x] for x in temp if x != '-']
    song_tag_idx[song] = temp_idx
    
    
with open('./song_tags.pkl', 'wb') as f:
    pickle.dump(song_tag, f)
    
with open('./song_tags_idx.pkl', 'wb') as f:
    pickle.dump(song_tag_idx, f)
```

# **3. FastText**

```python
from gensim.models import FastText

model = FastText(songs_vec, vector_size=100, window=5, min_count=1, workers=4, sg=1)
```

```python
def similar_songs(title_song):
    idx = song2idx[title_song]
    result = model.wv.most_similar(idx)
    similar_list = []
    for song_idx, cosine in result:
        temp = [idx2song[song_idx],round(cosine,4)]
        similar_list.append(temp)
        print(temp)
        
    return similar_list
```

```python
def search_song(name, data=songs_ind):
    data['match'] = data['singer'].apply(lambda x: x if name in x else 'X')
    sub_data = data[data['match']!='X'][['title','singer','final_tag']]
#     print(f'총 {sub_data.shape[0]}개의 곡이 검색되었습니다.')
    
    return sub_data    
```
